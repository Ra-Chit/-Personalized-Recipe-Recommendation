```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from a CSV file
df = pd.read_csv("E:\PROJECTS\FOOD RECOMD\indian_food.csv")

# Replace -1 with NaN to represent missing values
df.replace(-1, pd.NA, inplace=True)

# Data preprocessing
df.dropna(inplace=True)
df["ingredients"] = df["ingredients"].str.lower().str.strip()
df[["prep_time", "cook_time"]] = df[["prep_time", "cook_time"]].apply(pd.to_numeric, errors="coerce")


user_input = input("Enter the ingredients you have (comma-separated): ")
user_ingredients = [ingredient.strip().lower() for ingredient in user_input.split(",")]

# Filter the dataset based on user's input ingredients
filtered_df = df[df["ingredients"].apply(lambda x: any(ingredient in x for ingredient in user_ingredients))]

# Prompt the user to select the flavor profile
print("Select the desired flavor profile:")
flavor_profiles = df["flavor_profile"].unique()
for i, profile in enumerate(flavor_profiles, 1):
    print(f"{i}. {profile}")
flavor_choice = int(input("Enter the number corresponding to the flavor profile:1.Sweet 2.Spicy 3.Bitter "))
user_flavor_profile = flavor_profiles[flavor_choice - 1]

# Prompt the user to select the diet
print("Select the desired diet:")
diets = df["diet"].unique()
for i, diet in enumerate(diets, 1):
    print(f"{i}. {diet}")
diet_choice = int(input("Enter the number corresponding to the diet:1.VEG 2.NON VEG "))
user_diet = diets[diet_choice - 1]

# Filter the dataset based on user's input ingredients, flavor profile, and diet
filtered_df = filtered_df[
    (filtered_df["flavor_profile"] == user_flavor_profile) &
    (filtered_df["diet"] == user_diet)
]

# Define weights for ranking factors
ingredient_weight = 0.5
flavor_weight = 0.3
diet_weight = 0.2

# Calculate relevance score for each dish
filtered_df["relevance_score"] = (
    filtered_df["ingredients"].apply(lambda x: sum(ingredient in x for ingredient in user_ingredients)) * ingredient_weight +
    (filtered_df["flavor_profile"] == user_flavor_profile).astype(int) * flavor_weight +
    (filtered_df["diet"] == user_diet).astype(int) * diet_weight
)

# Sort the dataset based on relevance score in descending order
ranked_df = filtered_df.sort_values(by="relevance_score", ascending=False)

# Define the number of top recommendations to display
num_recommendations = 5

# Select the top recommended dishes from the ranked dataset
top_recommendations = ranked_df.head(num_recommendations)

# Print the top recommended dishes
print()
print("User Inputs:")
print("Ingredients:", ", ".join(user_ingredients))
print("Flavor Profile:", user_flavor_profile)
print("Diet:", user_diet)
print()
if top_recommendations.empty:
    print("Sorry, no matching suggestions.")
else:
    # Print the top recommended dishes
    print("Top", num_recommendations, "Recommended Dishes:")
    for index, row in top_recommendations.iterrows():
        print("Dish:", row["name"])
        print("Ingredients:", row["ingredients"])
        print("Preparation Time:", row["prep_time"], "minutes")
        print("Cooking Time:", row["cook_time"], "minutes")
        print("Flavor Profile:", row["flavor_profile"])
        print("Diet:", row["diet"])
        print("Course:", row["course"])
        print("State:", row["state"])
        print("Region:", row["region"])
        print("----------")
```

Data Vizualization

```python
plt.figure(figsize=(8, 6))
plt.bar(top_recommendations["name"], top_recommendations["relevance_score"])
plt.xlabel("Dish")
plt.ylabel("Relevance Score")
plt.title("Top Recommended Dishes")
plt.xticks(rotation=45)
plt.show()
```

```python
from collections import Counter

all_ingredients = df["ingredients"].str.split(", ")
ingredient_counts = Counter([ingredient for ingredients in all_ingredients for ingredient in ingredients])
popular_ingredients = ingredient_counts.most_common(10)

plt.figure(figsize=(8, 6))
plt.bar([ingredient[0] for ingredient in popular_ingredients], [ingredient[1] for ingredient in popular_ingredients])
plt.xlabel("Ingredients")
plt.ylabel("Count")
plt.title("Popular Ingredients")
plt.xticks(rotation=45)
plt.show()
```
